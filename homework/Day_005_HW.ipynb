```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = '../data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
app_train
```

```python
app_train.columns
```

```python
#資料0~30區間的選取
app_train2 = app_train.loc[:30, ['SK_ID_CURR', 'TARGET', 'CODE_GENDER', 'DAYS_EMPLOYED', 'AMT_ANNUITY', 'DAYS_REGISTRATION', 'AMT_REQ_CREDIT_BUREAU_DAY']]
app_train2
```

```python
app_train2.head(15)
```

```python
app_train2.cov()
```

```python
#Summary statistics on DAYS_REGISTRATION
app_train2['DAYS_REGISTRATION'].describe()
```

```python
#平均數
app_train2['DAYS_REGISTRATION'].mean()
```

```python
#標準差
app_train2['DAYS_REGISTRATION'].std()
```

由註冊天數欄位的描述值觀察，標準差與最小值兩個差距非常大且有負值，這部分資料可能會是干擾或誤植

```python
import matplotlib.pyplot as plt
%matplotlib inline
app_train2['DAYS_REGISTRATION'].hist()
plt.xlabel('DAYS_REGISTRATION')
```
